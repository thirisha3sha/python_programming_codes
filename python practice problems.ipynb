```python
#perfect number 6,28
def perfect(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    return sum==n
n=int(input("enter the number:"))
res=perfect(n)
if(res):
    print("perfect number!")
else:
    print("not perfect number!!")
```

```python
#prime number 7,11,13,17,19
def prime_num(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True
n=int(input("enter the number:"))
if(prime_num(n)):
    print("prime number!")
else:
    print("not a prime number!")
```
