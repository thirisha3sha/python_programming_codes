**FASCINATING NUMBER:**
A number is called fascinating if after multiplying it by 2 and 3 and concatenating the results with the original number, all digits from 1 to 9 appear exactly once.

Example:
Consider 192

192 × 1 = 192

192 × 2 = 384

192 × 3 = 576

Concatenated: 192384576

It contains all digits from 1 to 9 exactly once → ✅ Fascinating!

````python
def facinating(n):
  if n<100:
    return False
  concat=str(n)+str(n*2)+str(n*3)
  digits=sorted(concat)
  num_list=['1','2','3','4','5','6','7','8','9']
  return digits==num_list
n=int(input())
if facinating(n):
  print("fascinating num!")
else:
  print("not a fascinating num!")
````

For example, 6 is a perfect number

GCD(GREATEST COMMON DIVISIOR )/ HCF

GCD of two numbers is the largest number that divides both numbers exactly (without a remainder).

````python
def gcd(a,b):
  while b!=0:
    temp=a
    a=b
    b=temp%b
  return a
a=int(input())
b=int(input())
print("GCD:",gcd(a,b))
````

````python
class gcd:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def find(self):
    while self.b!=0:
      temp=self.a
      self.a=self.b
      self.b=temp%self.b
    return self.a
a=int(input())
b=int(input())
print("gcd:",gcd(a,b).find())

````

Output: true
Explanation: The input string contains all characters from ‘a’ to ‘z’.

Problem3:Two strings s1 and s2 are called isomorphic if there is a one-to-one mapping possible for every character of s1 to every character of s2. And all occurrences of every character in ‘s1’ map to the same character in ‘s2’.
Examples:

Input:  s1 = “aab”, s2 = “xxy”
Output: True
Explanation: ‘a’ is mapped to ‘x’ and ‘b’ is mapped to ‘y’.

Input:  s1 = “aab”, s2 = “xyz”
Output: False
Explanation: One occurrence of ‘a’ in s1 has ‘x’ in s2 and other occurrence of ‘a’ has ‘y’.

````python
def isomorphic_check(s1,s2):
  if len(s1)!=len(s2):
    return False
  mapping={}
  used_chars=set()
  for ch1,ch2 in zip(s1,s2):
    if ch1 in mapping:
      if mapping[ch1]!=ch2:
        return False
    else:
      if ch2 in used_chars:
        return False
      mapping[ch1]=ch2
      used_chars.add(ch2)
  return True
print(isomorphic_check("aab","xxy"))
print(isomorphic_check("aab","xyz"))

````

example 1: inpu1:12332 o/p: returns 1
example 2: inp1:251532 o/p: returns 3
example 3: inp1:10101 o/p: returns -1 # as no digit to removed

````python
def is_palindrome(s):
  return s==s[::-1]
def find_digit_to_remove(n):
  n=str(n)
  left=0
  right=len(n)-1
  while left<right:
    if n[left]!=n[right]:
      if is_palindrome(n[:left]+n[left+1:]):
        return int(n[left])
      elif is_palindrome(n[:right]+n[right+1:]):
        return int(n[right])
      else:
        return "not possible"
    left+=1
    right-=1
  return -1

print(find_digit_to_remove(1010230))
print(find_digit_to_remove(1111231))
print(find_digit_to_remove(251532))
print(find_digit_to_remove(10101))
print(find_digit_to_remove(12332))
print(find_digit_to_remove(23324))

````
